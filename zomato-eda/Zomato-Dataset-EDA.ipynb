# Zomato Dataset Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpt
import seaborn as sns
%matplotlib inline
```

```python
#Set figure size for all viz to 12x6 grid
mpt.rcParams['figure.figsize'] = (12, 6)
```

```python
df = pd.read_csv('datasets/zomato-dataset/zomato.csv', encoding = 'latin-1')
df.head()
```

```python
df.columns
```

```python
df.info() #To check datatypes, use df.dtypes
```

```python
#Only numerical features are emitted out by describe()
df.describe()
```

```python
#Number of null values for each feature(column)
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap='viridis')
```

```python
df_country = pd.read_excel('datasets/zomato-dataset/country-code.xlsx')
df_country.head()
```

```python
#Merge dataframes on specific feature, similar to sql joins
final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
final_df.head(2)
```

```python
#Country names sorted by record count in the dataset
country_names = final_df.Country.value_counts().index
```

```python
#Count of records for each country
country_counts = final_df.Country.value_counts().values
```

```python
#Pie Chart - Top 3 countries by Zomato usage.
plt.pie(country_counts[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

#### Observation: As per the graph, Zomato's maximum business is the Indian market. Low overseas adoption boils down to the popularity of other restaurant review apps (ex. Yelp) which have had a longer history in USA and Europe.

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
ratings
```

```python
sns.barplot(x = "Aggregate rating", y = "Rating Count", data = ratings)
```

#### Observations:
   1. Count for Not Rated (Agg Rating = 0.0) is very high
   2. Maximum ratings lie between 2.8 to 3.9

```python
#Find number of restaurants having 0 rating by country
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index().rename(columns = {0 : 'Rating Count'})
```

```python
#Find currency for each country
final_df[['Country', 'Currency']].drop_duplicates().sort_values(['Country'])
```

```python
# Top 5 cities with most restaurants on Zomato
city_names = final_df.City.value_counts().index

city_counts = final_df.City.value_counts().values

plt.pie(city_counts[:5], labels = city_names[:5], autopct = '%1.1f%%')
```

```python
#Top 10 cusisines on Zomato
cuisine_names = final_df.Cuisines.value_counts().index

cuisine_counts = final_df.Cuisines.value_counts().values

plt.pie(cuisine_counts[:10], labels = cuisine_names[:10], autopct = '%1.1f%%')
```
